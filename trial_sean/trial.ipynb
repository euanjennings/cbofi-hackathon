```python
import zipfile
import os
from IPython.display import HTML

# Files to include in the zip
files = [
    'app.py',
    'requirements.txt',
    'runtime.txt',
    'risk_ai_data_v3_3.csv',
    'fraud_classifier.pkl'
]

zip_name = 'fraud_app_deploy.zip'

with zipfile.ZipFile(zip_name, 'w', zipfile.ZIP_DEFLATED) as zipf:
    for file in files:
        if os.path.exists(file):
            zipf.write(file)
            print(f'✅ Added: {file}')
        else:
            print(f'❌ Missing: {file}')

print(f'\nZip created: {zip_name}')
```
