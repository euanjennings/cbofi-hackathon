**Load data via SQL**

<sub>Note: schemas may either be "dsp" or "api", this can be found on the dataset documentation</sub>

```python
database = connect_to_database()

q = """
  SELECT *
  FROM api."<insert_table_name>"
  LIMIT 10
"""
database.execute(q)

#Show query result as dataframe
cols = [desc[0] for desc in database.description]
df = pd.DataFrame(database.fetchall(), columns=cols)
df.head()
```
